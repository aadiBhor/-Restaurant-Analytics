# Task 1 : Top Cuisines
1.Determine the top three mostcommon cuisines in the dataset.             
2.Calculate the percentage ofrestaurants that serve each of the top cuisines.



#### 1) Top CuisinesDetermine the top three mostcommon cuisines in the dataset. 

```python
all_cuisines =df['Cuisines'].str.split(",").explode().str.strip()
top_cuisines =all_cuisines.value_counts().head(3).reset_index()
print("Top Three Most Common cuisines : ")
top_cuisines
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
sns.barplot(data=top_cuisines,x='Cuisines',y='count',hue='Cuisines',legend=True, palette='Set2')
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.title('Top 3 Most Common Cuisines')
plt.tight_layout()
plt.show()
```

Insight: The top three common cuisines are North Indian, Chinese, and fast food.

#### 2) Calculate the percentage of restaurants that serve each of the top cuisines.

```python
df['Cuisine_list']=df['Cuisines'].str.split(",")
top_cuisines = all_cuisines.value_counts().head(3).index.tolist()

Total_restaurants=df['Restaurant ID'].nunique()

cuisine_percentages={}
for i in top_cuisines:
    count = df['Cuisine_list'].apply(lambda x: i in x).sum()
    percentage = (count / Total_restaurants) * 100
    cuisine_percentages[i] = round(percentage, 2)

cuisine_percentages
```

```python
cuisines = list(cuisine_percentages.keys())+['other']
a,b,c=list(cuisine_percentages.values())
d=100-a-b-c
percentages = [a,b,c,d]


# Pie chart
plt.pie(percentages, labels=cuisines, autopct="%1.1f%%",shadow=True)
plt.title("Percentage of Restaurants Serving Top Cuisines")
plt.show()
```

Insight: Around 31% of restaurants serve North Indian, while Chinese (9%) and Fast Food (7%) are less common compared to other cuisines.

# Task 2:  City Analysis
1.Identify the city with the highest number of restaurants in the dataset.          
2.Calculate the average rating for restaurants in each city.        
3.Determine the city with the highestaverage rating         

#### 1) Identify the city with the highest number of restaurants in the dataset 

```python
city_counts = df['Restaurant ID'].groupby(df['City']).count().reset_index()
city_with_highest_num_of_restaurants=city_counts.sort_values(by='Restaurant ID' ,ascending=False).head(6)
city_with_highest_num_of_restaurants.head(1)
```

```python
sns.barplot(data=city_with_highest_num_of_restaurants,x='City', y="Restaurant ID", palette='viridis',hue='City')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.title('Number of Restaurants by City')
plt.grid( linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

Insight: New Delhi has the (5473) most restaurants in the dataset, so it is the top city .

#### 2) Calculate the average rating for restaurants in each city.

```python
avg_rating_city =df.groupby('City')['Aggregate rating'].mean().reset_index()
sorted_avg_rating = avg_rating_city.sort_values(by='Aggregate rating' ,ascending=False).head(20)
sorted_avg_rating
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

sns.barplot(data=sorted_avg_rating, x='City', y='Aggregate rating',hue='City', palette='coolwarm')
plt.xticks(rotation=90, ha='right')
plt.xlabel('City')
plt.ylabel('Average Aggregate Rating')
plt.title('Average Restaurant Rating by City Top')
plt.tight_layout()
plt.show()
```

Insight: citys like Inner City, Quezon City, Makati City, and Pasig City have the highest average restaurant ratings

```python
avg_rating_city =df.groupby('City')['Aggregate rating'].mean().reset_index()
sorted_avg_rating = avg_rating_city.sort_values(by='Aggregate rating' ,ascending=False)
plt.figure(figsize=(20,6))
sns.lineplot(data=sorted_avg_rating, x='City', y='Aggregate rating', marker='o')
plt.xticks(rotation=90, ha='right')
plt.xlabel('City')
plt.ylabel('Average Aggregate Rating')
plt.title('Average Restaurant Rating by City ')
plt.tight_layout()
plt.show()
```

Insight: Most cities have good restaurant ratings around 3.5 to 5, with a few cities showing exceptionally high satisfaction and some needing improvement.

#### 2) Calculate the percentage of restaurantsin each price range category

# Task 4 : Online Delivery
1.Determine the percentage of restaurants that offer online delivery.        
2.Compare the average ratings of restaurants with and without online delivery

#### 1) Determine the percentage of restaurants that offer online delivery.  

```python
Total_restaurants = len(df)
offer_online_delivery = len(df[df['Has Online delivery']=='Yes'])
percentage_offer_online_delivery = round((offer_online_delivery/Total_restaurants)*100,2)
not_offer_online = 100-percentage_offer_online_delivery
X={'Yes':percentage_offer_online_delivery,'No':not_offer_online}

plt.pie(x=X.values(), labels=X.keys() ,autopct='%1.1f%%')
plt.title("percentage of restaurantsin each price range category ")
plt.tight_layout()
plt.show()
```

Insight: Only 26% of restaurants offering online delivery & 74% of restaurants Not offering online delivery.

#### 2) Compare the average ratings of restaurants with and without online delivery

```python
a=df.groupby('Has Online delivery')['Aggregate rating'].mean()
a
```

```python
plt.figure(figsize=(7,6))
sns.barplot(data=df,x='Has Online delivery',y='Aggregate rating',hue='Has Online delivery',legend=True,errorbar=('ci', False),palette='viridis')
plt.title('Average ratings of restaurants with and without online delivery')
plt.grid( linestyle="--", alpha=0.6)
plt.yticks(np.arange(0, 4.5, 0.5))
plt.tight_layout()
plt.show()
```
